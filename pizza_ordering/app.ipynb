```python
# Pizza Class 
class Pizza:
    def __init__(self, description, cost):
        self._description = description
        self._cost = cost
        
    def get_description(self):
        return self._description
    
    def get_cost(self):
        return self._cost
  
# Pizza çeşitlerini, içeriklerini ve fiyatlarını barındıran pizza alt sınıfları
class KlasikPizza(Pizza):
    def __init__(self):
        description = "Klasik pizza, domates sosu ve peynirle hazirlanmistir."
        cost = 49.99
        super().__init__(description, cost)
    

class MargaritaPizza(Pizza):
    def __init__(self):
        description = "Margarita pizza, domates sosu, mozarella peyniri ve taze feslegen yapraklariyla hazirlanmistir."
        cost = 54.99
        super().__init__(description, cost)
    

class TurkPizza(Pizza):
    def __init__(self):
        description = "Turk pizzasi, sucuk, sosis, biber, domates, kasar peyniri ve zeytinle hazirlanmistir."
        cost = 67.99
        super().__init__(description, cost)
    

class SadePizza(Pizza):
    def __init__(self):
        description = "Sade pizza, pepperoni, ve mozarella peyniriyle hazirlanmistir."
        cost = 59.99
        super().__init__(description, cost)

# Decarator sınıfı
class Decorator(Pizza):
    def __init__(self, pizza):
        self._pizza = pizza
        
    def get_description(self):
        return self._pizza.get_description()
    
    def get_cost(self):
        return self._pizza.get_cost()

# Pizzanın üstüne koyacak malzemelerin fiyatları ve açıklamalarını barındıran malzeme sınıfları
class Zeytin(Decorator):
    def __init__(self, pizza):
        description = pizza.get_description() + " Zeytin eklenmistir."
        cost = pizza.get_cost() + 5.0
        super().__init__(Pizza(description, cost))

class Mantar(Decorator):
    def __init__(self, pizza):
        description = pizza.get_description() + " Mantar eklenmistir."
        cost = pizza.get_cost() + 5.0
        super().__init__(Pizza(description, cost))

class KeciPeyniri(Decorator):
    def __init__(self, pizza):
        description = pizza.get_description() + " Keci peyniri eklenmistir."
        cost = pizza.get_cost() + 5.0
        super().__init__(Pizza(description, cost))

class Et(Decorator):
    def __init__(self, pizza):
        description = pizza.get_description() + " Et eklenmistir."
        cost = pizza.get_cost() + 5.0
        super().__init__(Pizza(description, cost))

class Sogan(Decorator):
    def __init__(self, pizza):
        description = pizza.get_description() + " Sogan eklenmistir."
        cost = pizza.get_cost() + 5.0
        super().__init__(Pizza(description, cost))

class Misir(Decorator):
    def __init__(self, pizza):
        description = pizza.get_description() + " Misir eklenmistir."
        cost = pizza.get_cost() + 5.0
        super().__init__(Pizza(description, cost))
```

```python
def main():
    # Ekrana menü yazdırma
    print("/***** MENÜ *****/")
    print("1. Klasik Pizza - 49.99 TL")
    print("2. Margarita Pizza - 54.99 TL")
    print("3. Türk Pizza - 67.99 TL")
    print("4. Sade Pizza - 59.99 TL")
    print("5. Çıkış")

    # Kullanıcı tarafından pizzanın klavyedeki 1-5 arası sayılar ile istediği pizzayı seçmesi
    secim_pizza = int(input("Lütfen bir pizza seçin (1-5 arası sayılar): "))
    while secim_pizza < 1 or secim_pizza > 5:
        secim_pizza = int(input("Geçersiz seçim. Lütfen bir pizza seçin (1-5 arası sayılar): "))

    # Seçilen pizzanın fiyatını hesapla
    match secim_pizza:
        case 1:
            pizza = KlasikPizza()
        case 2:
            pizza = MargaritaPizza()
        case 3:
            pizza = TurkPizza()
        case 4:
            pizza = SadePizza()
        case 5:
            return 'Pizza siparişi iptal edildi.'
    
    # Kullanıcının malzeme seçmesi
    print("Lütfen bir malzeme seçin:")
    print("1. Zeytin - 5 TL")
    print("2. Mantar - 5 TL")
    print("3. Keçi Peyniri - 5 TL")
    print("4. Et - 5 TL")
    print("5. Soğan - 5 TL")
    print("6. Mısır - 5 TL")
    print("7. Hiçbir sos istemiyorum")
    secim_sos = int(input("(1-7 arası sayılar): "))
    while secim_sos < 1 or secim_sos > 7:
        secim_sos = int(input("Geçersiz seçim. Lütfen bir malzeme seçin (1-7 arası sayılar): "))

    # Seçilen malzemeyle beraber toplam fiyatın hesaplanması
    match secim_sos:
        case 1:
            sos = Zeytin(pizza)
        case 2:
            sos = Mantar(pizza)
        case 3:
            sos = KeciPeyniri(pizza)
        case 4:
            sos = Et(pizza)
        case 5:
            sos = Sogan(pizza)
        case 6:
            sos = Misir(pizza)
        case 7:
            sos = pizza
            
    toplam_fiyat = '{:.2f}'.format(sos.get_cost())

    # Kullanıcıya sipariş özeti ve fiyat bilgilendirmesi

    print(f'Seçtiğiniz pizza ve malzeme: {sos.get_description()}.')
    print(f'Toplam fiyat: {toplam_fiyat}.')
    # Kullanıcıdan isim, tc, kredi kartı ve şifresinin istenmesi
    isim = input("Lütfen isminizi giriniz: ")

    tc = int(input("Lütfen TC kimlik numaranızı giriniz: "))
    while tc <= 9999999999 or tc >= 100000000000:
        tc = int(input("Lütfen geçerli bir TC kimlik numaranızı giriniz: "))
    

    kart_numarasi = int(input("Lütfen kredi kartı numaranızı giriniz: "))
    while kart_numarasi <= 999999999999999 or kart_numarasi >= 10000000000000000:
        kart_numarasi = int(input("Lütfen geçerli bir kredi kartı numaranızı giriniz: "))
    

    kart_sifresi = int(input("Lütfen kredi kartı şifrenizi giriniz: "))
    while kart_sifresi <= 999 or kart_sifresi >= 10000:
        kart_sifresi = int(input("Lütfen geçerli bir kredi kartı şifrenizi giriniz: "))
    

    # Sipariş başarıyla oluşturulursa, oluşturulan zamanın alınması
    siparis_zamani = datetime.datetime.now()
    
    # Kullanıcı bilgilerinin bir database'e kaydedilmesi
    with open("Orders_Database.csv", "a", newline="", encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow([isim, tc, kart_numarasi, kart_sifresi, sos.get_description(), toplam_fiyat, siparis_zamani])
main()
```
